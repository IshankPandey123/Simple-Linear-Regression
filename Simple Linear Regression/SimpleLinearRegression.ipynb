```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('height-weight.csv')
```

```python
df.head()
```

```python
# Scatter plot
plt.scatter(df['Weight'] , df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
```

```python
# Correlation
df.corr()
```

```python
# Seaborn for visulaization
import seaborn as sns
sns.pairplot(df)
```

```python

df.head()
```

```python
# Independent and dependent feature
X = df[['Weight']] # independent feature should be data frame or 2 dimensional array
y = df['Height'] # this variable can be in series or 1D Array
```

```python
# Train , Test
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.25 , random_state = 42)
```

```python
X_train.shape
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# Apply Simple Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression(n_jobs = -1)
```

```python
regression.fit(X_train , y_train)
```

```python
regression
```

```python
print("Coefficient or slope : " , regression.coef_)
print("Intercept : " , regression.intercept_)
```

```python
# plot Training data best fit line
plt.scatter(X_train , y_train)
plt.plot(X_train , regression.predict(X_train))
```

Prediction on Test Data
1. predicted height output = Intercept + coef_(weight)
2. y_pred_test = 156.420 + 17.29(X_test)

```python
# Prediction for test_data
y_pred = regression.predict(X_test)
```

```python
# Performance Metrics
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
mse = mean_squared_error(y_test , y_pred)
mae = mean_absolute_error(y_test , y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

R Square

Formula

R^2 = 1 - SSR/SST

R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test , y_pred)
```

Adjusted R2=1-[(1-R2)^ * (n - 1) / (n - k - 1) ]

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python
# Display Adjusted R - squared
1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
```
